```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris['data']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(iris['data'],iris['target'],test_size=0.2,stratify = iris['target'], random_state=25)
```

```python
from sklearn.metrics import classification_report
```

## 결정트리

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt.score(x_train, y_train)
y_pred_tree = dt.predict(x_test)
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors = 3)
kn.fit(x_train, y_train)
kn.score(x_train, y_train)
y_pred_knn = kn.predict(x_test)
print(classification_report(y_test, y_pred_knn))
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C = 12, max_iter = 1000)
lr.fit(x_train, y_train)
lr.score(x_train, y_train)
y_pred_logic = lr.predict(x_test)
print(classification_report(y_test, y_pred_logic))
```
